```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

```python
spaceship.drop(['PassengerId','Destination','Name'],axis=1, inplace=True)
```

```python
spaceship.isnull().sum()
```

```python
spaceship.dropna(inplace=True)
```

```python
spaceship.isnull().sum()
```

```python
spaceship['Cabin']=spaceship['Cabin'].apply(lambda x: x.split('/')[0])
```

```python
features= spaceship.drop('Transported',axis=1)
features_encoded = pd.get_dummies(features)


```

```python
target= spaceship['Transported']
```

```python
features_encoded
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled= scaler.fit_transform(features_encoded)
```

- Now let's use the best model we got so far in order to see how it can improve when we fine tune it's hyperparameters.

```python

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)
```

```python

bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    bootstrap=True, # Set to False for Pasting
    random_state=42
)
bagging.fit(X_train, y_train
)
print("Bagging Accuracy:", bagging.score(X_test, y_test))
```

```python

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
print("Random Forest Accuracy:", rf.score(X_test, y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gb.fit(X_train, y_train)
print("Gradient Boosting Accuracy:", gb.score(X_test, y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
ada.fit(X_train, y_train)
print("AdaBoost Accuracy:", ada.score(X_test, y_test))
```

**Grid/Random Search**

For this lab we will use Grid Search.

- Define hyperparameters to fine tune.

```python

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False]
}
```

```python

# Initialize model
rf = RandomForestClassifier(random_state=42)
```

```python

# Grid search
from sklearn.model_selection import GridSearchCV


grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
```

```python

# Best model and accuracy
from sklearn.metrics import accuracy_score


best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

- Run Grid Search

```python

print("Best Parameters:", grid_search.best_params_)
print("Test Set Accuracy:", accuracy)
```
